```python
import os
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
root_path = os.path.dirname(os.getcwd())

clean_data_path = root_path+'/data/cleaned'

with open(clean_data_path+'/job_description_cleaned.txt', 'r') as f:
    jd = f.readlines()

jd = " ".join(jd)
print(jd)
```

```python
resume_data = pd.read_csv(clean_data_path+'/resume_cleaned.csv', index_col=False)
resume_data.head()
```

```python
resume_texts = [text for text in resume_data['clean_texts']]
print(resume_texts)
```

```python
corpus = [jd]
corpus += resume_texts
len(corpus)
```

```python
vectorizer = TfidfVectorizer()
doc_matrix = vectorizer.fit_transform(corpus)
```

```python
cosine_similarity(doc_matrix[0:1], doc_matrix)
```
